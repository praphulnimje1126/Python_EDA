# Q2- Data Preperation

1 Dropping irrelevant columns and rows

```python
data_set.drop(['Length', 'Speed','Type','Height restriction', 'Model', 'Height','Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section','Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle','Drop', 'Soft opening date', 'Fast Lane available', 'Replaced','Track layout', 'Fastrack available', 'Soft opening date.1','Closing date','Replaced by', 'Website','Flash Pass Available', 'Must transfer from wheelchair', 'Theme','Single rider line available', 'Restraint Style','Flash Pass available', 'Acceleration', 'Restraints', 'Name','speed1', 'speed2', 'speed1_value', 'speed1_unit','height_value', 'height_unit'],inplace=True,axis=1)
data_set
```

```python
data_set.dtypes
```

```python
data_set['opening_date_clean']
data_set['opening_date_clean']= pd.to_datetime(data_set['opening_date_clean'])
data_set['opening_date_clean']
```

```python
data_set['year_introduced']=pd.to_numeric(data_set['year_introduced'])
data_set['year_introduced']
```

```python
data_set.rename(columns={'coaster_name':'Coaster_Name','year_introduced':'Year_Introduced',
                         'opening_date_clean':'Opening_Date','speed_mph':'Speed_mph',
                         'height_ft':'Height_ft'},inplace=True)
data_set.columns
```

```python
data_set.isnull().sum()
```

# 3. Feature Understanding
(Univariable analysis)

Plotting Feature Distributions

1. Histogram

```python
set1=data_set['Year_Introduced'].value_counts().head(10).plot(kind='bar',title='Top Years Coasters Introduced')
set1.set_xlabel('Year')
set1.set_ylabel('Count')
```

```python
set2=data_set['Speed_mph'].plot(kind='hist',bins=20,title='Coaster Speed (mph)')
set2.set_xlabel('Speed (mph)')
```

```python
set2=data_set['Speed_mph'].plot(kind='kde',title='Coaster Speed (mph)')
set2.set_xlabel('Speed (mph)')
```
