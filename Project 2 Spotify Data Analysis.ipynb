# **Q-2 Top 10 most popular songs which has popularity more than 90**

```python
most_popularity=df_tracks.query('popularity>90',inplace=False).sort_values('popularity',ascending=False).head(10)
most_popularity 
```

```python
df_tracks.set_index('release_date',inplace=True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
df_tracks[['artists']].iloc[18]
```

```python
df_tracks['duration']=df_tracks['duration_ms'].apply(lambda x: round(x/1000))
df_tracks.drop(columns=['duration_ms'],inplace=True)
df_tracks.head()
```

```python
corr_df=df_tracks.drop(columns=['key','mode','explicit'],axis=1).corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt='.1g',vmin=-1,vmax=1,center=0,cmap='plasma',linewidth=1,linecolor='black')
heatmap.set_title("correlation heatmap between variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90) 
```

```python
sample_df=df_tracks.sample(int(0.004*len(df_tracks)))
len(sample_df)
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y='loudness',x='energy',color='lime').set(title='regression plot between loudness and energy')
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y='popularity',x='acousticness',color='skyblue').set(title='regression plot between popularity and acousticness')
```

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks['dates']=pd.to_datetime(df_tracks['dates'])
year=df_tracks.dates.dt.year
```

```python
sns.displot(year,discrete=True,aspect=2,height=5,kind='hist').set(title='number of songs per year')
```

```python
total_dr=df_tracks.duration
fig,ax=plt.subplots(figsize=(18,7))
fig=sns.barplot(x=year,y=total_dr,ax=ax).set(title='year wise song duration')
plt.xticks(rotation=90)
plt.show()
```

```python
total_dr=df_tracks.duration
fig,ax=plt.subplots(figsize=(10,5))
fig=sns.lineplot(x=year,y=total_dr,ax=ax).set(title='year wise duration of song')
plt.xticks(rotation=60)
plt.show()
```
